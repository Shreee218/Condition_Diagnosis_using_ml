```python
symptoms = []

for col in data.columns[1:]:
    symptoms.append(data[col].unique())

symptoms
```

```python
unique_symp = []

for row in symptoms:
    for symp in row:
        if (symp not in unique_symp) and (symp is not np.nan):
            unique_symp.append(symp.strip())
        else:
            print('existing', end="\r")
```

```python
values = [0.0] * 394
symp_dict = dict(zip(unique_symp, values))
symp_dict
```

```python
len({key: 0 for key in symp_dict})
```

```python
encoded = []

for i in range(len(data)):
    row = data.iloc[i].values
    temp = dict(zip(unique_symp, values))

    for i in row:
        if i is not np.nan:
            temp[i.strip()] = 1.0

    encoded.append(temp)
```

```python
pre_process = pd.DataFrame(data=encoded)
pre_process['disease'] = data['Disease']
pre_process.fillna(value=0,inplace=True)
pre_process
```

```python
pre_process.to_csv('pre_processed.csv')
```

```python
x=pre_process.iloc[:,:-1].values
y=pre_process.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1000)
```
